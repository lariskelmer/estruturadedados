```python
%pip install nltk
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import regex as re

%pip install sortedcontainers
from sortedcontainers import SortedSet

MAX_OUTPUT_CHARS = 100000
```

```python
#### CARREGADO DE: https://www.corpusdata.org/portuguese.asp

with open('./sample_data/text.txt', 'r') as file:
    corpus_text = file.read(MAX_OUTPUT_CHARS)

print(corpus_text)
```

```python
def get_unique_words(text: str) -> set:
  """
  Obtém as palavras únicas do texto.
  Args:
      text (str): Texto a ser processado.
  Returns:
      set: Conjunto de palavras únicas.
  """

  words = text.split()  # divide o texto em palavras
  unique_words = set(words)  # cria um conjunto de palavras únicas

  return unique_words

def remove_hyphens(words_list):
  # Expressão regular para remover hífens no final das palavras
  pattern = r'^-+|-+$'
  without_hyphens = [re.sub(pattern, '', word) for word in words_list]
  return without_hyphens

def clear_text(text: str) -> str:
  """
  Limpa e normaliza o texto, retirando caracteres especiais, stop words (baseadas na biblioteca nltk)
      além de deixá-lo em minúsculas e excluir demarcações da base.
  Args:
      text(str): corpus usado como base.
  Returns:
      str: texto processado.
  """

  text_without_special_chars_and_punct = re.sub(r'[.,!?;:()_+*º°@/\\|-ª´€"«»%=`]', '', text) # remove pontuação
  text_without_digits = re.sub(r'\d+(?:\.\d+)?|\d+(?::\d+)?', ' ', text_without_special_chars_and_punct) # remove dígitos
  text_without_extra_blanks = re.sub(r'\s{2,}', ' ', text_without_digits) # remove espaços em branco extras
  text_lower = text_without_extra_blanks.lower() # deixa em minúsculo
  unique_words = get_unique_words(text_lower)

  words = text_lower.split() # separa em palavras - retorna uma lista

  words_without_trailing_hyphens = remove_hyphens(words)

  stop_words = set(nltk.corpus.stopwords.words('portuguese')) # seta stop words da nltk
  extra_words = ['ão', 'ó', 'wwwangodebatesblogspotcom', 'wwwombembwablogspotcom', 'uolcombrvyaestelarplasm', 'balumukenohiphop', 'balumukenuhiphop']
  stop_words.update(extra_words) #se quiser entrar com palavras novas, mantém essa linha e adicionar: extra_words=['palavra1', 'palavra2', ...]

  filtered_words = [word for word in words_without_trailing_hyphens if word not in stop_words] # remove stop words

  cleared_text = ' '.join(filtered_words) # junta a lista em texto corrido
  return cleared_text

clean_text = clear_text(corpus_text)
print(clean_text)
```

```python
def create_avl_tree(text: str):
  unique_words = text.split()
  avl_tree = SortedSet(unique_words)
  return avl_tree

avl_tree = create_avl_tree(clean_text)
# Imprime o conteúdo da Árvore AVL (os elementos da árvore)
for word in avl_tree:
    print(word)
```
